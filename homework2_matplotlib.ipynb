```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# values for sigma:
sigma1 = 0.5
sigma2 = 1.5
sigma3 = 2.5
```

```python
# the x-axis values to plot will be:
x_axis = np.linspace(-10, 10, 1000)
```

```python
# Gaussian function using our different Sigmas
tempCalc1 = np.exp(-0.5*(x_axis/sigma1)**2)
tempCalc2 = sigma1*(2*np.pi)**(1/2)
y_axis1 = tempCalc1/tempCalc2
tempCalc1 = np.exp(-0.5*(x_axis/sigma2)**2)
tempCalc2 = sigma2*(2*np.pi)**(1/2)
y_axis2 = tempCalc1/tempCalc2
tempCalc1 = np.exp(-0.5*(x_axis/sigma3)**2)
tempCalc2 = sigma3*(2*np.pi)**(1/2)
y_axis3 = tempCalc1/tempCalc2
```

```python
#Ploteando:
plt.plot (x_axis, y_axis1, linestyle="-", label="$\sigma=0.5$", color="red")
plt.plot (x_axis, y_axis2, linestyle=":", label="$\sigma=1.5$", color="green")
plt.plot (x_axis, y_axis3, linestyle="--", label="$\sigma=2.5$", color="orange")

#Titles and legend:
plt.title("Gaussian function for different $\sigma$ values"), plt.xlabel('x-values'), plt.ylabel("g(x)"), plt.legend(), plt.grid(), plt.show()
#plt.savefig("funciongaussian.png", dpi=300)
```
